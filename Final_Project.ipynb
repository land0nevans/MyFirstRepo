#####I found the path in my google drive for my world_cities.csv file and imported this as a dataframe named 'df'.

```python
df = pd.read_csv('drive/MyDrive/population_data/data/world_cities.csv')
```

#####I created a new 'pop_M' column by using df[] = , placing the name 'pop_M' in the square brackets and the conversion on the other side of the equals, which was dividing the raw population number by 1 million.


```python
df["pop_M"] = df["pop"]/1000000
```

I then removed the 'pop' column from the entire dataframe, by reassigning df = df.drop(columns=["pop"]) (this was easily explained in the pandas documentation, using the 'columns' parameter.)

```python
df=df.drop(columns=["pop"])
```

My name begins with an 'L', so I filtered the original dataframe, keeping only rows where the 'city' starts with "L".
I found df.str.startswith() in a stack overflow reply, and I was able to use df["city"] to only apply the rule to the 'city' column. To check this worked correctly, I created a new variable 'L_cities' and headed this.
#####( https://stackoverflow.com/questions/17957890/pandas-select-from-dataframe-using-startswith )

```python
L_cities = df[df["city"].str.startswith("L")]
L_cities.head(6)
```

######I created a new variable to print the city at the top of the list, by using iloc[0] to select the first row. I printed this new variable to check this worked correctly.

```python
first_city= L_cities.iloc[0]
print(first_city)
```

#####Then, I created new variable 'france_cities' by filtering my dataframe, only selecting data where the original 'country' column is equal (==) to the 'country' value in my first_city variable, which is France.

```python
france_cities = df[df["country"] == first_city["country"]]
```

Then, for the final output i created new variable 'top5' and sorted my new 'france_cities' dataframe using 'DataFrame.nlargest(n, columns)' from the pandas documentation, with n being the number of rows and columns being the label "pop_M".

```python
top5 = france_cities.nlargest(5, "pop_M")
```

For my final output, I printed a message using my created variables.

```python
print("My name begins with L, so i selected city:")
print(first_city)
print()
print("This city is in France, and the 5 largest countries in France are:")
print(top5)
```
